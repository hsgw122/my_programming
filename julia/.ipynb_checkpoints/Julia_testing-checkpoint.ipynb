エージェントの基本データ。設定値と、記録値で分けた。再定義にはstructだけではエラーを吐く。module内で整理すれば置き換えてくれる。

```julia
module MY
#モジュール内容
#
#SVOAgent_Data :エージェントの定義パラメータ
# NAME
# angle
# self_w
# other_w
# strategy
#SVOAgent_Memory :選択頻度など、行動から得た結果をまとめたデータ
# p_A
# fre
# Mat_cor
# svo_can
# score
# Hist
#Matrix_Legasy :使うマトリックスのデータ。[posi ,m_number]で、前半が1～8番目、後半が何番目のマトリックスかを指定
# matrix
#score_AVE :scoreの平均を出す


  #戦略名
  p_X = 10
  p_M = 11
  p_svo = 12

  struct SVOAgent_Data #変更しないパラメータ
    NAME
    angle::Float64
    self_w::Float64
    other_w::Float64
    strategy
  end

  #宣言の方法
  function SVOAgent_Data(_NAME,_angle,_strategy)
    NAME = _NAME
    angle = _angle
    strategy = _strategy
    return SVOAgent_Data(NAME,angle,cos(pi / 180 * angle),sin(pi / 180 * angle),strategy)
  end

  mutable struct SVOAgent_Memory #変更可能なパラメータ
    #記録される内容

    #自分にとっての相手の選択確率
    p_A

    #AかBかの選択頻度の記録
    fre::Array{Int64,1} #1次元配列
    #マトリックスに対応する選択の記録、大本のマトリックスのインデックスで検索する。辞書型
    Mat_cor
    #SVOの記録、計算してから、候補をプラスしていく。0から275までの8段階
    svo_can::Array{Int64,1}

    #獲得スコアのリスト。
    score::Array{Float64,1}

    #SVO予測のヒストグラムリスト
    Hist::Array{Float64,2}
  end

  #宣言
  function SVOAgent_Memory(game_n,hist_size)
    score = zeros(Float64,game_n)
    Hist = zeros(Float64,8,hist_size)
    return SVOAgent_Memory(0.5,[0,0],Dict(),zeros(Int64,8),score,Hist)
  end

  #使うマトリックスの保存先、変更不可。定義時にのみ変更可能
  struct Matrix_Legasy
    matrix::Array{Float64,2} #2次元配列、要素が実数。予定としては、8×60000の2次元配列
  end

  #宣言、マトリックスゲームの一覧
  function Matrix_Legasy(game_n)
    return Matrix_Legasy(rand(8,game_n))
  end

  #テスト宣言用。同じゲームで
  function Matrix_Legasy_test(game_n)
    tes_lega = zeros(Float64,8,game_n)
    println(tes_lega)
    for i = 1:game_n
      tes_lega[:,i] = [2.0,2.0,0.0,3.0,3.0,0.0,1.0,1.0]
    end
    return Matrix_Legasy(tes_lega)
  end

  #平均を出す
  function score_AVE(Game_Memory)
    score = Game_Memory.score
    return sum(score)/length(score)
  end

end
```

```julia
function play_sub(self_w,other_w,p_A,matrix)
  row_A = (matrix[1] * self_w + matrix[2] * other_w) * p_A + (matrix[3] * self_w + matrix[4] * other_w) * (1 - p_A)
  row_B = (matrix[5] * self_w + matrix[6] * other_w) * p_A + (matrix[7] * self_w + matrix[8] * other_w) * (1 - p_A)
  row = row_A - row_B
  #println(row)
  if row >= 0
    return 0
  else
    return 1
  end
end

function play(agent_Data,agent_memory,matrix)
  #println("play")
  #println(agent_Data.angle)
  #println(matrix[8])
  self_w = agent_Data.self_w
  other_w = agent_Data.other_w
  p_A = agent_memory.p_A
  #row_A = (matrix[1] * self_w + matrix[2] * other_w) * p_A + (matrix[3] * self_w + matrix[4] * other_w) * (1 - p_A)
  #row_B = (matrix[5] * self_w + matrix[6] * other_w) * p_A + (matrix[7] * self_w + matrix[8] * other_w) * (1 - p_A)
  #row = row_A - row_B
  #println(row)
  #if row >= 0
    #return 0
  #else
    #return 1
  #end

  return Int64(play_sub(self_w,other_w,p_A,matrix))

end

play(a,b,[2,2,0,3,3,0,1,1])
```

マトリックスから角度を出す関数。マトリックスと確率で出力される。

```julia
#マトリックスと確率から角度を出す
function toArcTan(matrix,p_A)
    row_over =  matrix[1] * p_A + matrix[3] * ( 1 - p_A ) - matrix[5] * p_A - matrix[7] * ( 1 - p_A )
    row_under = matrix[2] * p_A + matrix[4] * ( 1 - p_A ) - matrix[6] * p_A - matrix[8] * ( 1 - p_A )
    _tan = -( row_over ) / ( row_under )
    if row_over == 0 && row_under == 0
      _tan = -1
    end
    
    #println("_tan",_tan)
    #dを付ければ、角度で扱う。でなければラジアンとなる
    arc_ = atand(_tan)
    if arc_ < 0
      arc_ += 360
    end
    if row_under < 0
      arc_ += 180
    end
    
    return arc_
  end
  
  test_at = toArcTan([2,2,0,3,3,0,1,1],0.5)
  println(test_at)
  
  #8等分に考える
  function slice8(atn)
    #浮動小数点にしてから整数型に変換
    slicer = atn / 45
    #println(slicer)
    #slicer = slicer / 45
    #println(slicer)
    slicer_f = floor(slicer)
    #println(slicer_f)
  
    #ちょうどかどうか
    just_flag = 0
  
    if slicer_f - slicer == 0
      just_flag = 1
    end
    return Int64(slicer_f) % 8,just_flag
  end
  
  #角度を360°から8の範囲に分解する
  test_slice_8 = test_at / 45
  println(test_slice_8)
  
  println(slice8(test_at))
  
  for i = 1:10
    println(slice8((i-1)*45))
  end
  
  using Random
  
  temoo = 1000
  
  stac_moo_list = zeros(Int64,8,1)
  #seed値の設定
  Random.seed!()
  moo = rand(Float64,8,temoo)
  for i = 1:temoo
    moo_i = moo[:,i]
    moo_arc = toArcTan(moo_i,0.5)
    moo_slice = slice8(moo_arc)
    #println(moo_arc)
    #println(moo_slice)
    stac_moo_list[moo_slice[1]+1] += 1
  end
  
  println(stac_moo_list)
  
  valuation = 4
  stac_moo_list = zeros(Int64,8,1)
  for a = 1:valuation
    for b = 1:valuation
      for c = 1:valuation
        for d = 1:valuation
          for e = 1:valuation
            for f = 1:valuation
              for g = 1:valuation
                for h = 1:valuation
                  moo_i = zeros(Float64,8,1)
                  #println(moo_i)
                  moo_i = [Float64(a),b,c,d,e,f,g,h]
                  #println(moo_i)
                  moo_arc = toArcTan(moo_i,0.5)
                  #println(moo_arc)
                  moo_slice = slice8(moo_arc)
                  #println(moo_arc)
                  #println(moo_slice)
                  stac_moo_list[moo_slice[1]+1] += 1
                end
              end
            end
          end
        end
      end
    end
  end
  
  println(stac_moo_list)
```

```julia
# 4つの候補に1プラスする関数
function canPlus(memory,matrix,ch,round_)
    #println()
  
    atn =  toArcTan(matrix,memory.p_A)
    #println(atn)
    mas_tan8 = slice8(atn)
    #println(mas_tan8)
  
    list = zeros(Float64,8)
    #println(list)
  
    #Aを選んだ場合
    if ch == 0
      #println(mas_tan8[2])
      #ぴったりなら
      if mas_tan8[2] == 1
        #println("just")
        #println("+")
        #println("mas_tan8[1] ",mas_tan8[1])
        index_tan = mas_tan8[1]
        index_tan = index_tan % 8 + 1
        list[index_tan] += 1
        memory.svo_can[index_tan] += 1
        #println(memory.svo_can)
      #ぴったりでない場合は
      else
        #println("over")
      end
  
      dbu = 0
  
      #ノーマルは、1と4
      for i = mas_tan8[1] + 1 + dbu : mas_tan8[1] + 4 + dbu
        #println("+")
        #println(i)
        index_tan = i
        index_tan = index_tan % 8 + 1
        list[index_tan] += 1
        memory.svo_can[index_tan] += 1
  
        #println(memory.svo_can)
      end
  
    #Bを選んだ場合
    else
      dbu = 0
  
      #ノーマルは5と8
      for i = mas_tan8[1] + 5 + dbu :mas_tan8[1] + 8 + dbu
        #println("+")
        #println(i)
        index_tan = i
        index_tan = index_tan % 8 + 1
        list[index_tan] += 1
        memory.svo_can[index_tan] += 1
        #println(memory.svo_can)
      end
  
    end
  
    #println(list)
    if round_ == 1
      memory.svo_can = list
    else
      for i = 1:length(list)
        #memory.svo_can[i] = memory.svo_can[i] * list[i]
      end
    end
    #println(memory.svo_can)
    #println()
  
  end
```

```julia
round = 2

#2回目の対戦

#使うマトリックスの更新
A_Mas = Mas.matrix[:,round]
B_Mas = AB_change(A_Mas)

println(A_Mas)
println(B_Mas)


#戦略によって分岐し、確率を推定する。

function update_p_(player,memory,Mas,my_player,my_memory)
  p_A = 0.5

  #println(player.NAME)
  strategy_ = my_player.strategy

  #選択頻度の場合
  if strategy_ == MY.p_X
    #println("p_X")
    #println(memory.p_A)
    p_A = memory.fre[1]/sum(memory.fre)
  end

  #マトリックスとの対応を見る場合
  if strategy_ == MY.p_M
    #println("p_M")
    if get(memory.Mat_cor,Mas,9) != 9
      #println("true")
      cho = memory.Mat_cor[Mas]
      #println(cho)
      if cho == 0
        p_A = 1.0
      else
        p_A = 0.0
      end
    else
      p_A = 0.5
    end
  end

  #SVOから推定する場合
  if strategy_ == MY.p_svo
    #println("p_svo")
    svo_ch = zeros(Float64,8)
    svo_leg = memory.svo_can
    #println(svo_leg)
    svo_leg = svo_leg/sum(svo_leg)
    for i = 1:8
      if play_sub(cosd(45*(i - 1)),sind(45*(i - 1)),p_A,Mas) == 0
        svo_ch[i] = 1.0 * svo_leg[i]
      else
        svo_ch[i] = 0.0 * svo_leg[i]
      end
    end

    #println(svo_ch)
    #println(sum(svo_ch))
    p_A = sum(svo_ch)
  end

  #p_A = 0.5

  #println(p_A)
  my_memory.p_A = p_A
  #println(my_memory.p_A)
end

update_p_(A_player,A_memory,A_Mas,B_player,B_memory)
update_p_(B_player,B_memory,B_Mas,A_player,A_memory)
```
